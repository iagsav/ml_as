### 1. Визуализируйте корреляции между признаками, находящимися в наборе данных (heatmap). 

```python
data_heatmap = df.drop(['Release date', 'Price'], axis=1).corr()
sns.heatmap(data_heatmap, cmap="YlOrRd", square=True)
plt.show()
```

### 2. Вычислите корреляцию Пирсона и Спирмена между Price и Max (Min) Resolution. Сравните полученные величины корреляции. Постройте гистограммы для указанных признаков.

```python
print(df[['Price', 'Max resolution', 'Low resolution']].corr(method='pearson'))
print(df[['Price', 'Max resolution', 'Low resolution']].corr(method='spearman'))
spearPears = pd.concat([df[['Price', 'Max resolution', 'Low resolution']].corr(method='pearson'), 
                        df[['Price', 'Max resolution', 'Low resolution']].corr(method='spearman')])
spearPears.plot.bar()
plt.show()
```

### 3. Постройте график изменения средней цены на камеры по годам.

```python
mean_price = df.groupby('Release date')['Price'].mean()
mean_price.plot.bar()
plt.show()
```

### 4. Определите компанию, камеры которой наиболее часто встречаются в наборе данных.

```python
kort = df['Model'].to_string().split('\n')
start_dict = collections.defaultdict(int)

for k in kort:    
    start_dict[re.search('[A-z]+', k).group(0)] += 1  
    
max_value = max(start_dict.values())
final_dict = {k: v for k, v in start_dict.items() if v == max_value}      
       
final_dict
```
