## 1. Удалите столбцы, ценность которых для оценки стоимости машины низка.

```python
ad = ad.drop(['dateCreated','postalCode','lastSeen'],axis = 1)
ad
```

## 2.Удалите повторяющиеся строки, строки содержащие пропуски в данных

```python
ad = ad.drop_duplicates()
ad = ad.dropna()
ad
```

```python
ad.shape
```

## 3.Удалите данные в столбцах, выходящие за некоторые пределы (год регистрации, цена, мощность двигателя) 

```python
ad = ad[ad.yearOfRegistration > 2000] 
ad = ad[(ad.price > 5000) & (ad.price < 7500)] 
ad = ad[ad.powerPS < 100]
```

```python
ad.shape
```

## 4.Заполнить пропущенные данные в строковых полях.



```python
ad = ad.fillna(0)
ad
```

## 5.Постройте гистограмму по маркам автомобилей, типам кузова и используемому топливу.

### 5.1 Гистограмма по маркам авто

```python
plt.figure(figsize=(3, 8))
plt.title('Марки авто')
sns.countplot(data=ad, y='brand')
plt.show()
```

### 5.2 Гистограмма по типам кузова

```python
plt.figure(figsize=(3, 8))
plt.title('Тип кузова')
sns.countplot(data=ad, y='vehicleType')
plt.show()
```

### 5.3 Гистограмма по используемому топливу

```python
plt.figure(figsize=(3, 8))
plt.title('Топливо')
sns.countplot(data=ad, y='fuelType')
plt.show()
```

## 6. Добавьте в данные новый признак, который представляет собой длину названия автомобиля.

```python
def get_len_name(name):
    return str(len(name.split('\n')[0].strip()))


len_name = ad['name'].apply(get_len_name)
ad['lenName']=len_name
ad
```

## 7.Постройте карту корреляций между признаками.

```python
sns.heatmap(ad.corr(), annot=True, fmt='.1f', linewidths=.5)
plt.plot()
plt.show()
```

## 8.Определите коэффициент корреляции Пирсона и Спирмена между стоимостью автомобиля и километражом. Оцените значение p-value. Постройте плотность распределения признаков. 

```python
print(ad[['price','kilometer']].corr(method='pearson'))
print(ad[['price', 'kilometer']].corr(method='spearman'))
```
