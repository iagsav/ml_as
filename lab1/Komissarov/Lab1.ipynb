### 2. Определите какой части пассажиров удалось выжить. Посчитайте долю выживших пассажиров.


```python
from math import isnan
import matplotlib.pyplot as plt
import pandas as panda

df = panda.read_csv('Data/test.csv')

survive = 0
total = df.Name.count()

for element in df.Cabin:
    if isinstance(element, float) and isnan(element):
        continue
    else:
        survive += 1;

print('Ratio of survivers and total passengers = {:.7}'.format(survive / total))

labels = ['Survivers', 'Victims']
people = [survive, total]

fig1, ax1 = plt.subplots()
ax1.pie(people, labels=labels, autopct='%1.4f%%',
        shadow=True, startangle=60)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show();
```

### 3. Какую долю пассажиры первого класса составляли среди всех пассажиров?


```python
import pandas as panda

df = panda.read_csv('Data/test.csv')
firstClass = df[df['Pclass'] == 1]['PassengerId'].count()

people = [firstClass, total]
labels = ['First Class', 'Other Classes']

print('Ratio of first class passengers and total class passengers = {:.7}'.format(firstClass / total))
fig1, ax1 = plt.subplots()
ax1.pie(people, labels = labels,colors=[ (0.9, 0.9, 0.9),  (0.2, 0.7, 0.2)], autopct='%1.4f%%', shadow=True, startangle=180);
ax1.axis('equal')
plt.show()
```
