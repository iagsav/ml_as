### 1. Постройте графики количества осадков по годам в разных штатах. Сделайте выводы по построенным графикам.

```python
array = tuple(data_frame['SUBDIVISION'].unique())

for item in array:
    data_frame_1 = data_frame[['SUBDIVISION','YEAR','ANNUAL']]
    data_frame_1 = data_frame_1[data_frame_1['SUBDIVISION'] == item]  
    data_frame_1.plot(x = 'YEAR', y = 'ANNUAL', title = str(item))
    plot.show()
```

### 2. Постройте графики количества осадков по месяцам по всем годам кумулятивно. Сделайте выводы по построенным графикам.

```python
for i in range(2, 14):
    data_frame_2 = data_frame.iloc[:, [1, i]]
    res2 = 0
    lst21 = []
    lst22 = []
    for j in range(data_frame_2[data_frame_2.columns[0]].min(), (data_frame_2[data_frame_2.columns[0]].max()) + 1):
        data_frame_22 = data_frame_2.copy()
        data_frame_22 = data_frame_22[data_frame_22['YEAR'] == j]
        res2 += data_frame_22[data_frame_22.columns[1]].sum()
        lst21.append(j)
        lst22.append(res2)
    tup21 = ('YEAR', lst21)
    tup22 = ('OVERALL', lst22)
    tmp2 = dict([tup21, tup22])
    df2tmp = pandas.DataFrame(tmp2)
    df2tmp.plot(x = 'YEAR', y = 'OVERALL', title = str(i-1) + ") " + str(data_frame_2.columns[1]))
    plot.show()
```

### 3. Постройте графики количества осадков по штатам, используйте boxplot. Сделайте выводы по построенным графикам.

### 4. Определите штаты, для которых количество осадков минимально и максимально.

```python
data_frame_4 = data_frame.groupby('SUBDIVISION')['ANNUAL'].sum().sort_values()
print("Minimum - " + str(data_frame_4.index[0]) + ", maximum - " + str(data_frame_4.index[len(data_frame_4) - 1]))
```
