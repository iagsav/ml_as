## Удалите столбцы, ценность которых для оценки стоимости машины низка. ##

```python
table.drop(['dateCrawled','offerType', 'abtest', 'fuelType','lastSeen','postalCode','dateCreated'], axis = 1)
```

## Удалите повторяющиеся строки, строки содержащие пропуски в данных.

```python
table = table.drop_duplicates()
table = table.dropna()
table
```

## Удалите данные в строках, выходящие за некоторые пределы (год регистрации, цена, мощность двигателя).

```python
table = table.drop(table[table.yearOfRegistration < 2000].index)
table = table.drop(table[table.price < 2000].index)
table = table.drop(table[table.powerPS > 100].index)
table.shape
```

## Заполнить пропущенные данные в строковых полях.

```python
table = table.fillna(0)
table
```

## Постройте гистограмму по маркам автомобилей, типам кузова и используемому топливу.

```python
plt.figure(figsize=(8, 10))
plt.title('Auto brand')
sns.countplot(data=table, y='brand')
plt.show()
```

## Добавьте в данные новый признак, который представляет собой длину названия автомобиля.

```python
def getLenthName(name):
    return str(len(name.split('\n')[0].strip()))

lenthName = table['name'].apply(getLenthName)
table['lenthName']=lenthName
table
```

## Постройте карту корреляций между признаками.

```python
sns.heatmap(table.corr())
```
