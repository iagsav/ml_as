Вычислите корреляцию Пирсона и Спирмена между признаками Gender и Missions. Сравните полученные величины корреляции. Постройте гистограммы для указанных признаков. Оцените значение p-value.

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

lbE = LabelEncoder()
df_copy=df[['Gender']].join(df['Missions'].str.split(',', expand=True).stack().reset_index(drop=True, level=1).rename('Missions').map(lambda x: x.strip()))

df_copy["Gender"] = lbE.fit_transform(df_copy["Gender"])
df_copy["Missions"] = lbE.fit_transform(df_copy['Missions'].astype(str))
pearson_correlation, pearson_pvalue = stats.pearsonr(df_copy['Gender'], df_copy['Missions'])
print('Pearson:')
print('Correlation - ', pearson_correlation)
print('p-value - ', pearson_pvalue)

spearman_correlation, spearman_pvalue = stats.spearmanr(df_copy['Gender'], df_copy['Missions'])
print('\nSpearman:')
print('Correlation - ', spearman_correlation)
print('p-value - ', spearman_pvalue)
```

Значение корреляции в обоих случаях близко к нулю, следовательно корреляция между двумя рассматриваемыми признаками очень слабо выражена.
Значение p-value ниже уровня значимости (0.05) в обоих случаях, значит мы отклоняем нулевую гипотезу.
